```python
#n으로 출력을 받기
n = input()

#0으로만 이루어진 리스트를 만들자
list_0 = n.split('1')
list_0 = sorted(list_0)
for i in range(len(list_0)):
  if list_0[0] != '':
    break
  else:
    list_0.remove( '')


#1로만 이루어진 리스트를 만들자
list_1 = n.split('0')
list_1 = sorted(list_1)

for i in range(len(list_1)):
  if list_1[0] != '':
    break
  else:
    list_1.remove( '')

#이제 len을 통해서 리스트에 가장 적게 할당된 값을 구하고 이걸 답으로 출력
if len(list_0) < len(list_1) :
  print(len(list_0))
if len(list_0) > len(list_1) :
  print(len(list_1))
if len(list_0) == len(list_1) :
  print(len(list_1))
```

2. 모두 구해서 set하기---> 중복제거

3. 이후 sort로 순차대로 정렬

4. 반복문을 사용해서 다음 숫자에 안이어지는 부분을 찾기

###풀이
랜덤으로 중복이 안되는 뽑기를 사용해서 뽑아서 값을 구하자.

```python
#조합 라이브러리 가저오기
import random
import math
import itertools
combine_list = []

# 숫자 받기
n = int(input())
list_num = list(map(int,input().split()))

#조합으로 경우의수 돌리기
for i in range(n):
  k = itertools.combinations(list_num,i)
  for i in k:
    a = list(i)
    b = sum(a)
    combine_list.append(b)
#set함수로 중복 제거
combine_list = set(combine_list)
combine_list = list(combine_list)
#리스트를 돌려서 순서가 안맞는 부분을 찾기
for i in range(len(combine_list)):
  if combine_list[i] + 1 != combine_list[i+1]:
    print(combine_list[i] + 1)
    break  
```
